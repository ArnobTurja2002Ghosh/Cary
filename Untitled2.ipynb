```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns


import ipywidgets as widgets
from ipywidgets import interact, interact_manual

df = pd.read_csv('open_gym.csv')
```

```python
df['total']=df['total_females']+df['total_males']
df['year']=df['open_gym_start'].str[0:4]
df['start_hour']=df['open_gym_start'].str[11:19]
df['end_hour']=df['open_gym_end'].str[11:19]
df['day']=df['open_gym_end'].str[8:10].astype(int)
df['month']=df['open_gym_end'].str[5:7].astype(int)
df['year']=df['open_gym_end'].str[0:4].astype(int)
df['majority_sex'] = df.apply(lambda x:'M' if(x['total_males']>x['total_females']) else('F' if(x['total_males']<x['total_females']) else('N')), axis=1)
```

```python
@interact(FacilityTitle=['All','Bond Park Community Center', 'Herbert C. Young Community Center',
       'Middle Creek Community Center', 'Cary Arts Center'])
def f(FacilityTitle):
  if(FacilityTitle=="All"):
    df1= df.groupby(['open_gym_activity', 'group'])['total'].mean().reset_index()
  else:
    df1=df[df['facility_title']==FacilityTitle].groupby(['open_gym_activity', 'group'])['total'].mean().reset_index()
  df_hm = df1.pivot(index='open_gym_activity', columns='group', values='total')#.fillna(0)
  #df_hm1=df_hm[['SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']]
  sns.heatmap(data=df_hm, cmap='rocket')
```

```python
df3 = df[df['total_females'].notna() & df['total_males'].notna()]
# @interact(channel=(['Swimming', 'Wrestling', 'Fencing', 'Athletics', 'Shooting', 'Cycling', 'Rowing', 'Gymnastics', 'Combined']))
def select_sport(Location, kde, display_data):
    if(Location!='Combined'):
        df3_athletics=df3[(df3['location']==Location)]
        
    elif(Location=='Combined'):
        df3_athletics=df3[df3['location'].isin(['Sycamore Gymnasium', 'Magnolia Gymnasium', 'North Gym (Auxiliary)', 'Coach Kay Yow Court', 'South Gym'])]
    if(not kde):
        ax=sns.histplot(df3_athletics, x="total_residents", y="total_non_residents", binwidth=5, cmap='YlOrBr')
        if(display_data):
            sns.scatterplot(df3_athletics, x="total_residents", y="total_non_residents", hue="majority_sex", hue_order=['M', 'F', 'N'], ax=ax)
    elif(kde):
        fig, ax = plt.subplots()
        sns.kdeplot(df3_athletics, x="total_residents", y="total_non_residents", fill=True, ax=ax)
        if(display_data):
            sns.scatterplot(df3_athletics, x="total_residents", y="total_non_residents", hue="majority_sex", hue_order=['M', 'F', 'N'], ax=ax)

    ax.set(xlim=(0, 80), ylim=(0, 35), ylabel='total non-residents', xlabel='total residents')
interact(select_sport, Location = ['Combined', 'Sycamore Gymnasium', 'Magnolia Gymnasium', 'North Gym (Auxiliary)', 'Coach Kay Yow Court', 'South Gym'],
            kde = False,
            display_data = False);
```
